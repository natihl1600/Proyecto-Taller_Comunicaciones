```python
# Se cargan las bibliotecas de pandas, numpy y matplotlib
import pandas as pd 
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
import matplotlib. pyplot as plt

#bibliotecas extra para tests especificos 
from scipy.stats import anderson, normaltest, probplot, t
#from outliers import smirnov_grubbs as grubbs

print("Bibliotecas importadas correctamente")
```

```python
#Lectura de los datos de excel 
datos = pd.read_excel('Dataset.xlsx')
BPM = datos ['BPM']
```

Utilizando la funcion de anderson() se determina si los datos poseen una distribucion normal con un nivel de significancia α = 0.05
Se selecciona este valor por convencion 

```python
anderson_test = anderson(BPM, dist='norm') 
estadistico = anderson_test.statistic
crit_vals = anderson_test.critical_values
sig_levels = anderson_test.significance_level  
desviacion_estandar = np.std(BPM)
promedio = np.mean(BPM)

print("Estadistico A²:", estadistico)
print("Valores criticos:", crit_vals)
print("Niveles de significancia:", sig_levels)
print("Desviacion estandar:", desviacion_estandar)
print("Promedio:", promedio)

#Utilizando un significance level de 5% entonces el valor critico de interes es el de indice 2, puesto que los valores criticos tienen el orden [cv_15, cv_10, cv_5, cv_2_5, cv_1]
critico = crit_vals[2]

if estadistico > critico: 
    print("Los datos NO tienen una distribucion normal")
else:
    print("Los datos SI tienen una distribucion normal")
```

```python
#Grafico de probabilidad normal
plt.figure(figsize=(6, 6))
probplot(BPM, dist="norm", plot=plt)
plt.title("Probability Plot - Comparación con distribución normal")
plt.xlabel("Valores teóricos (Normal)")
plt.ylabel("Pulsaciónes por minuto")
plt.grid(True)
plt.show()
```

Del grafico anterior se puede observar como los datos se ajustan bastante bien a una curva normal

```python
#Histograma 
plt.figure()
plt.hist(BPM, bins=30, density=True, edgecolor='black')
plt.xlabel("Pulsaciones por minuto")
plt.ylabel("Frecuencia")
plt.title("Histograma de BPM")
plt.grid(True, alpha=0.3)
plt.show()
```

Del histrograma se puede observar que la mayoria de los datos se encuentran en la zona media, con una decadencia hacia los lados, como la curva normal de Gauss

```python
#Box plot y grubbs test

#grubbs.test(BPM, alpha=.05)

alpha = 0.05
n=len(BPM)
indice_mayor = np.argmax(np.abs(BPM - promedio))
G = np.abs(BPM[indice_mayor] - promedio) / desviacion_estandar

tcrit = t.ppf(1 - alpha/(2*n), df=n-2)
Gcrit = ((n - 1) / np.sqrt(n)) * np.sqrt((tcrit**2) / (n - 2 + tcrit**2))

if G > Gcrit:
    print("El test de Grubbs SI detectó un outlier")
else:
    print("El test de Grubbs No detectó un outlier")



plt.figure()
plt.boxplot(BPM, vert=True, patch_artist=True)
plt.ylabel("Pulsaciones por minuto")
plt.title("Box Plot de BPM")
plt.grid(True, axis='y', alpha=0.3)
plt.show()
```
